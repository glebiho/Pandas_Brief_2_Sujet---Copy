```python
#On veut afficher les produits dont la catégorie = boissons
df.loc[df['Categorie']=="Boissons"]
```

```python
df.loc[(df['Categorie']=="Boissons") & (df['Prix']>100)]
```

```python
#On veut afficher les produits dont la Catégorie = boissons et l'origine=CEE et prix > 100
df.loc[(df['Categorie']=="Boissons") & (df['Origine']=='CEE') & (df['Prix']>100)]
```

```python
#On veut afficher les produits dont la (Catégorie = boissons et origine = CEE) OU (catégorie = condiment)
df.loc[(df['Categorie']=="Boissons") & (df['Origine']=='CEE') | (df['Categorie']=='Condiments')]
```

```python
#On veut afficher les produits dont (Catégorie = viande ET origine = CEE) OU (catégorie = condiment ET origine = extérieur)
df.loc[(df['Categorie']=="Viande") & (df['Origine']=='CEE') | (df['Categorie']=='Condiments') & (df['Origine']=='Exterieur')]
```

```python
#On veut afficher les produits dont le prix > 70 et prix <=100
df.loc[(df['Prix']>70) & (df['Prix']<=100)]
```

```python
#Lister les aliments dont le prix est compris entre 100 et 200, et qui sont des « viandes »
df.loc[(df['Prix']>100) & (df['Prix']<200) & (df['Categorie']=="Viandes")]
```

```python
#Lister les 15 produits les moins chers
df.nsmallest(15, 'Prix')
```

```python
#Calculer la moyenne de prix des boissons distribuées à Lyon (pivot_table)
df1=df.loc[(df['Ville']=='Lyon')&(df['Categorie']=='Boissons')]
df1
df1.pivot_table(index=['Categorie'],columns=['Ville'],values=['Prix'],aggfunc=pd.Series.mean)
```

```python
#Quels sont les 5 produits les moins chers vendus à Lyon ?
df.loc[(df['Ville']=='Lyon')]
df.nsmallest(5, 'Prix')
```

```python
#En utilisant une boucle afficher la representation graphique de toutes les variables quantitative
import matplotlib.pyplot as plt
list_quan=['Prix', 'Stock']
plt.figure(figsize=(40,10))
i=1
for a in list_quan:
    plt.subplot(2,3,i)
    df[a].value_counts().plot(kind='hist',color='gray')
    i+=1
    plt.title(a)
```

```python
#En utilisant une boucle afficher la representation graphique de toutes les variables quanliative
list_qual=['Nom', 'Categorie', 'Ville', 'Origine']
plt.figure(figsize=(17,7))
i=1
for element in list_qual:
    plt.subplot(1,7,i)
    df[element].value_counts().plot(kind='pie')
    i+=1
    plt.title(element)
```
